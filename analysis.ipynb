```python
import pandas as pd
import numpy as np
import datetime
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use(['seaborn-whitegrid'])
plt.rcParams['font.family'] = 'AppleGothic'
```

# 연도별 기록 추이 및 분포 분석

```python
boston = pd.read_csv('data/boston_marathon_yearly.csv', index_col=0)
chicago = pd.read_csv('data/chicago_marathon_yearly.csv', index_col=0)
olympic = pd.read_csv('data/olympic_marathon_yearly.csv', index_col=0)
```

```python
olympic['중도포기 및 완주자'] = olympic['중도포기자'] + olympic['완주자']
olympic['중도포기 비율'] = olympic['중도포기자'] / olympic['중도포기 및 완주자']
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(15,5))
fig.suptitle('올림픽 마라톤 기록 추이')

sns.lineplot(ax=axes[0], x="year", y="금메달_기록", data=olympic)
sns.lineplot(ax=axes[1],x="year", y="상위10위_평균_기록", data=olympic)
sns.lineplot(ax=axes[2],x="year", y="중도포기 비율", data=olympic)

plt.savefig('img/olympic_trend.png')
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(15,5))
fig.suptitle('보스톤 마라톤 기록 추이')

sns.lineplot(ax=axes[0], x="year", y="금메달_기록", data=boston)
sns.lineplot(ax=axes[1],x="year", y="상위10위_평균_기록", data=boston)
sns.lineplot(ax=axes[2],x="year", y="상위500위_평균_기록", data=boston)

plt.savefig('img/boston_trend.png')
```

```python
fig, axes = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(15,5))
fig.suptitle('시카고 마라톤 기록 추이')

sns.lineplot(ax=axes[0], x="year", y="금메달_기록", data=chicago)
sns.lineplot(ax=axes[1],x="year", y="상위10위_평균_기록", data=chicago)
sns.lineplot(ax=axes[2],x="year", y="상위500위_평균_기록", data=chicago)

plt.savefig('img/chicago_trend.png')
```

```python
boston_indv = pd.read_csv('data/boston_marathon_individual.csv', index_col=0)
chicago_indv = pd.read_csv('data/chicago_marathon_individual.csv', index_col=0)
olympic_indv = pd.read_csv('data/olympic_marathon_individual.csv', index_col=0)
```

```python
olympic_fin = olympic_indv[olympic_indv['time']!='None'].reset_index()
olympic_fin['time'] = pd.to_datetime(olympic_fin['time'])
olympic_fin['time'] = [x.time() for x in olympic_fin['time'] ]
for i in range(len(olympic_fin)):
    minute = olympic_fin.loc[i, 'time'].hour * 60 + olympic_fin.loc[i, 'time'].minute
    olympic_fin.loc[i, 'minutes'] = minute
```

```python
fig, axes = plt.subplots(3, 3, sharex=True, sharey=False, figsize=(15,6))
fig.suptitle('올림픽 마라톤 기록 분포')


sns.histplot(ax=axes[0,0], data=olympic_fin[olympic_fin.year==2016], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[0,1], data=olympic_fin[olympic_fin.year==2012], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[0,2], data=olympic_fin[olympic_fin.year==2008], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[1,0], data=olympic_fin[olympic_fin.year==2004], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[1,1], data=olympic_fin[olympic_fin.year==2000], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[1,2], data=olympic_fin[olympic_fin.year==1996], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[2,0], data=olympic_fin[olympic_fin.year==1992], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[2,1], data=olympic_fin[olympic_fin.year==1988], x="minutes", hue="gender", element="poly")
sns.histplot(ax=axes[2,2], data=olympic_fin[olympic_fin.year==1984], x="minutes", hue="gender", element="poly")


axes[0,0].title.set_text('2016')
axes[0,1].title.set_text('2012')
axes[0,2].title.set_text('2008')
axes[1,0].title.set_text('2004')
axes[1,1].title.set_text('2000')
axes[1,2].title.set_text('1996')
axes[2,0].title.set_text('1992')
axes[2,1].title.set_text('1988')
axes[2,2].title.set_text('1984')

plt.show()

plt.savefig('img/olympic_hist.png')
```

# 날씨가 마라톤 경기에 미친 영향 분석

```python
olympic_after_40 = olympic[olympic['year']>1940]

fig, axes = plt.subplots(1, 3, sharex=True, sharey=False, figsize=(20,6))
fig.suptitle('1940년대 이후 기온에 따른 올림픽 마라톤 기록 분포')

sns.scatterplot(ax=axes[0], x="기온", y="금메달_기록", hue="year", style="성별", palette="Blues", s=100, legend=False, data=olympic_after_40)
sns.scatterplot(ax=axes[1], x="기온", y="상위10위_평균_기록", hue="year", style="성별", palette="Blues", s=100, legend=False, data=olympic_after_40)
sns.scatterplot(ax=axes[2], x="기온", y="중도포기 비율", hue="year", style="성별",  palette="Blues", s=100, data=olympic_after_40)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

axes[0].title.set_text('금메달 기록')
axes[1].title.set_text('상위10위 평균 기록')
axes[2].title.set_text('중도포기 비율')

plt.show()
plt.savefig('img/olympic_temp.png')
```
